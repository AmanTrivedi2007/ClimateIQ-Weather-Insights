```python
print(df.dtypes)
print("\n")
print(df.info())
```

```python
print(df.describe())
```

```python
print("\n")

print("Max average temperature in Celsius:", df['MaxTemperature_C'].mean())
print("Min average temperature in Celsius:", df['MinTemperature_C'].mean())
print("Average humidity in percentage:", df['AvgHumidity_Pct'].mean())
print("Total rainfall in mm:", df['Rainfall_mm'].sum())
print("\n")

maxtemp = df.loc[df['MaxTemperature_C'].idxmax()]
print("Day with maximum temperature was:",maxtemp['Date'])
print("With the Maximum temperature of :", maxtemp['MaxTemperature_C'])
```

```python
df = pd.DataFrame(weather_data)
max_humidity = df.loc[df['AvgHumidity_Pct'].idxmax()]
print("Day with maximum humidity was:", max_humidity['Date'])
print("With the Maximum humidity of :", max_humidity['AvgHumidity_Pct'])
print("\n")
min_temp = df.loc[df['MinTemperature_C'].idxmin()]
print("Day with minimum temperature was:", min_temp['Date'])
print("With the Minimum temperature of :", min_temp['MinTemperature_C'])    
print("\n")
max_rainfall = df.loc[df['Rainfall_mm'].idxmax()]
print("Day with maximum rainfall was:", max_rainfall['Date'])
print("With the Maximum rainfall of :", max_rainfall['Rainfall_mm'])
print("\n")
```

```python
print("Plotting the weather data...")

plt.Figure(figsize=(12, 6))
plt.plot(df['Date'], df['MaxTemperature_C'], label='Max Temperature (C)')
plt.plot(df['Date'], df['MinTemperature_C'], label='Min Temperature (C)')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.legend()
plt.title('Daily Max and Min Temperatures in August 2025')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(df['Date'], df['Rainfall_mm'], color='orange', label='Rainfall (mm)')
plt.xlabel('Date')
plt.ylabel('Rainfall (mm)')
plt.title('Daily Rainfall in August 2025')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(df['Date'], df['AvgHumidity_Pct'], color ='green' , label='Avg Humidity (%)')
plt.xlabel('Date')
plt.ylabel('Avg Humidity (%)')
plt.title('Daily Average Humidity in August 2025')
plt.legend()
plt.show()
```
